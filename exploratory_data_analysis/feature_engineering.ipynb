### Creating time windows

#### Morning -> 6 - 11:59
#### Afternoon -> 12 - 17:59
#### Evening -> 18 - 23:59
#### Midnight -> 0 - 5:59

#### Minutes for midnight -> 0 <= midnight <= 359
#### Minutes for morning -> 359 < morning <= 719
#### Minutes for afternoon -> 719 < afternoon <= 1079
#### Minutes for evening -> 1079 < evening <= 1439

```python
filtered_dataset.dtypes
```

```python
def get_time_of_day(dataset_row):
    if dataset_row["minutes_of_day"] >= 0 and dataset_row["minutes_of_day"] <= 359:
        time_of_day = "midnight"
    
    elif dataset_row["minutes_of_day"] >= 360 and dataset_row["minutes_of_day"] <= 719:
        time_of_day = "morning"
    
    elif dataset_row["minutes_of_day"] >= 720 and dataset_row["minutes_of_day"] <= 1079:
        time_of_day = "afternoon"
    
    else:
        time_of_day = "evening"
    
    return time_of_day

filtered_dataset["time_of_day"] = filtered_dataset.apply(get_time_of_day, axis=1)
filtered_dataset.dtypes
```

```python
"""
Different types of browsers
1. Chrome
2. Safari
3. Mozilla
4. Opera
5. Firefox
6. Google
7. Presto
8. Trident
"""

def get_browser(dataset_row):
    if "chrome" in dataset_row["user_agent"].lower().split("/", 1)[1]:
        device = "chrome"
    elif "safari" in dataset_row["user_agent"].lower().split("/", 1)[1]:
        device = "safari"
    elif "firefox" in dataset_row["user_agent"].lower().split("/", 1)[1]:
        device = "firefox"
    elif "opera" in dataset_row["user_agent"].lower().split("/", 1)[1]:
        device = "opera"
    elif "google" in dataset_row["user_agent"].lower().split("/", 1)[1]:
        device = "chrome"
    elif "mozilla" in dataset_row["user_agent"].lower().split("/", 1)[1]:
        device = "firefox"
    elif "trident" in dataset_row["user_agent"].lower().split("/", 1)[1]:
        device = "trident"
    elif "presto" in dataset_row["user_agent"].lower().split("/", 1)[1]:
        device = "presto"
    else:
        device = "not_known"
        
    return device


filtered_dataset["browser"] = filtered_dataset.apply(get_browser, axis=1)
# ["user_agent"].str.split("/", 1).str[1]
filtered_dataset.dtypes
```

```python
"""
Different types of devices
1. Windows
2. Mac
3. Linux
4. Iphone
5. Android
6. Ipad
"""

def get_device(dataset_row):
    if "windows" in dataset_row["user_agent"].lower():
        device = "windows"
    elif "macintosh" in dataset_row["user_agent"].lower():
        device = "mac"
    elif "linux" in dataset_row["user_agent"].lower():
        device = "linux"
    elif "iphone" in dataset_row["user_agent"].lower():
        device = "iphone"
    elif "android" in dataset_row["user_agent"].lower():
        device = "android"
    elif "ipad" in dataset_row["user_agent"].lower():
        device = "ipad"
    else:
        device = "not_known"
        
    return device


filtered_dataset["device"] = filtered_dataset.apply(get_device, axis=1)
filtered_dataset.dtypes
```
